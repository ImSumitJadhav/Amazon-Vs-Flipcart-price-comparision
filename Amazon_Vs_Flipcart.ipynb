```python
# Null values are replaced by "Unknown"
final_df['brand_fk'] = final_df['brand_fk']. fillna('Unknown')
final_df['brand_az'] = final_df['brand_az']. fillna('Unknown')

final_df["retail_price_fk"].dropna(inplace=True)
final_df["discounted_price_fk"].dropna(inplace=True)
```

# **5.Use of NLP**

```python
list1=[]
for i in final_df["category_fk"]:
  list1.append(i.replace(">>",""))

final_df["category_fk"]=list1
final_df
```

```python
import re
```

```python
def remchar(string):
  str = string
  result = re.sub(r'[^a-zA-Z\s]',"",str)
  return result
```

```python
final_df['category_fk']=final_df['category_fk'].apply(lambda x:remchar(x))

final_df['category_az']=final_df['category_az'].apply(lambda x:remchar(x))
```

```python
final_df['category_fk'][1]
```

```python
list2=[]
for i in final_df['category_fk']:
  res = " ".join(i.split())
  res=res.replace(" ",",")
  list2.append(res)

final_df["tags"]=list2
```

```python
final_df
```

```python

final_df= final_df.drop(['category_fk',"brand_fk","category_az","brand_az"], axis=1)
final_df
```

```python
#import spacy liabrary
import spacy
#loading the english language small model of spacy
en = spacy.load('en_core_web_sm')
```

```python
import string
def text_process(msg):  #Removing punctuations from data
    nopunc =[char for char in msg if char not in string.punctuation or char==","]
    nopunc=''.join(nopunc)
    return nopunc

def stop_words(sent):#Removing stopwards of language english from data
    word =[word for word in sent if word not in stopwords]
    word=''.join(word)
    return word
```

```python
stopwords = en.Defaults.stop_words
```

```python
final_df['tags'] = final_df['tags'].apply(text_process)
final_df['tags'] = final_df['tags'].apply(stop_words)
```

```python
final_df.head()
```

```python
# Stemming of words
from nltk.stem import PorterStemmer

ps=PorterStemmer()

def stem(text):   #User define function for stemming
   Y=[]
   for i in text.split():
      Y.append(ps.stem(i)) 

   return " ".join(Y)


final_df['tags']=final_df['tags'].apply(stem)
```
